# Newtons Low Cooling

```python
period=120 #minutes
Tpred=[68] #degrees celcius
Tamb=25.4 #degrees celcius
dt=1 #We want temp change per minute
K=0.02
tpred=range(period)
```

```python
#Implement the model
for t in range(1,period):
    dT=-K*(Tpred[t-1]-Tamb)*dt
    Tpred_update=Tpred[t-1]+dT
    Tpred.append(Tpred_update)
    
print(Tpred)
```

```python
Tobs=[68,64,60.3,56.8,54.5,49.6,46.8,44.1,40.3]
tobs=[0,3,6,9,12,18,24,30,40]

import matplotlib.pyplot as plt
%matplotlib inline

plt.figure(figsize=(8,5))
plt.plot(tpred,Tpred,'--k')
plt.plot(tobs,Tobs)
#plt.plot(tobs,Tobs,'--k')
plt.plot(tobs,Tobs,'ok')
plt.xlabel('Time in minutes',size=16)
plt.ylabel('Temperature',size=16)
plt.show()
```
