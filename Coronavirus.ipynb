# Coronavirus COVID-19 Timeseries of Global Cases

#### The recent outbreak of Coronavirus demonstrates how quickly a threatening disease can spread around the world due to human travel and migration and it shows that we might not be well prepare for a pandemic.

#### The Center for Sytems Science and Engineering (https://systems.jhu.edu/research/public-health/ncov/ (Links to an external site.)) at John Hopkins University has generated a dashboard for real-time tracking of confirmed infected cases, recovered cases, and deaths. The data originates from different non-profit and governmental sources and a curated lists of cases by day is available through Github.

#### This program create a figure that shows the timeseries of total number of confirmed cases, recovered cases, and deaths.

```python
# Import modules
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#Load Coronavirus data from git hub
url1 = 'https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_19-covid-Confirmed.csv'
url2 = 'https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_19-covid-Deaths.csv'
url3 = 'https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_19-covid-Recovered.csv'

df1 = pd.read_csv(url1)
df2 = pd.read_csv(url2)
df3 = pd.read_csv(url3)
```

```python
#Get the total number of people confirmed, dead and recovered by Coronavirus for each day and save it to one data frame
df1_sum=df1.iloc[:,4:].sum()
df2_sum=df2.iloc[:,4:].sum()
df3_sum=df3.iloc[:,4:].sum()
df = pd.DataFrame([df1_sum,df2_sum,df3_sum])
df= df.T
```

```python
# Converting the dates to date type module
df_date=pd.to_datetime(df1.iloc[:,4:].columns)
```

```python
# Plot the Corona virus timeseries
plt.figure(figsize=(12,8))       
plt.plot(df_date,df.iloc[:,0],'--y')
plt.plot(df_date,df.iloc[:,1],'ro',markersize=4)
plt.plot(df_date,df.iloc[:,2],'-g')

#Add the title
plt.title('Coronavirus Timeseries',size=16)

#Add y label
plt.ylabel('People',size=14)

#Modify the axes
plt.xticks(rotation=45)
plt.yticks(np.arange(0, max(max(df.iloc[:,0]),max(df.iloc[:,1]),max(df.iloc[:,2])), 10000))

#Add the legend
plt.legend(('Infected', 'Deaths', 'Recovered'),
           loc='upper left')
plt.show()
```
